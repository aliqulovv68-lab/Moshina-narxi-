```python

import pandas as pd
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

def load_data(filepath):
    logger.info(f"Fayl yuklanmoqda: {filepath}")
    try:
        df = pd.read_csv(filepath)
        logger.info(f"Yuklandi. Qatorlar: {len(df)}")
        return df
    except Exception as e:
        logger.error(f"Xatolik: {e}")
        raise

filepath = r"C:\Users\Yusuf\Desktop\1-project\Data\raw\used_car_price_prediction_1M.csv"
df = load_data(filepath)
```

```python
logger.info("Data exploration boshlandi")

# Shakl (qator va ustunlar soni)
logger.info(f"Dataset shakli: {df.shape[0]} qator, {df.shape[1]} ustun")

# Bo'sh qiymatlar
missing = df.isnull().sum().sum()
if missing > 0:
    logger.warning(f"Bo'sh qiymatlar topildi: {missing} ta")
else:
    logger.info("Bo'sh qiymatlar yo'q")

# Dublikatlar
duplicates = df.duplicated().sum()
if duplicates > 0:
    logger.warning(f"Dublikat qatorlar: {duplicates} ta")
else:
    logger.info("Dublikat qatorlar yo'q")

logger.info("Data exploration tugadi")
```
